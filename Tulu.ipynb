```python
import numpy as np
from tensorflow.keras.layers import Layer, Input, LSTM, Dense, Embedding, Dropout, Bidirectional
from tensorflow.keras.models import Model
from tensorflow.keras import initializers
from sklearn.metrics import classification_report
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder

class SelfAttention(Layer):
    def __init__(self, attention_size, **kwargs):
        super(SelfAttention, self).__init__(**kwargs)
        self.attention_size = attention_size
        self.W = self.add_weight(name='att_weight', shape=(attention_size, attention_size),
                                 initializer=initializers.get('glorot_uniform'),
                                 trainable=True)
        self.b = self.add_weight(name='att_bias', shape=(attention_size,),
                                 initializer=initializers.get('zeros'),
                                 trainable=True)
        self.u = self.add_weight(name='att_context_vector', shape=(attention_size, 1),
                                 initializer=initializers.get('glorot_uniform'),
                                 trainable=True)

    def call(self, x, mask=None):
        # Step 1: Calculate the attention weights
        uit = tf.tanh(tf.tensordot(x, self.W, axes=1) + self.b)  # (batch_size, seq_len, attention_size)
        ait = tf.tensordot(uit, self.u, axes=1)  # (batch_size, seq_len, 1)
        ait = tf.squeeze(ait, -1)  # (batch_size, seq_len)

        # Step 2: Apply softmax to get normalized attention weights
        ait = tf.exp(ait)

        if mask is not None:
            ait *= tf.cast(mask, tf.float32)

        ait /= tf.cast(tf.reduce_sum(ait, axis=1, keepdims=True) + tf.keras.backend.epsilon(), tf.float32)

        # Step 3: Calculate the weighted average of the input sequence
        weighted_input = x * tf.expand_dims(ait, -1)  # (batch_size, seq_len, attention_size)
        output = tf.reduce_sum(weighted_input, axis=1)  # (batch_size, attention_size)

        return output

    def compute_mask(self, inputs, mask=None):
        return None

    def compute_output_shape(self, input_shape):
        return (input_shape[0], input_shape[-1])


def build_bilstm_with_attention(vocab_size, embedding_dim, max_seq_len, num_classes):
    inputs = Input(shape=(max_seq_len,))
    embeddings = Embedding(vocab_size, embedding_dim)(inputs)
    bilstm = Bidirectional(LSTM(128, return_sequences=True))(embeddings)
    attention = SelfAttention(attention_size=256)(bilstm)
    dropout = Dropout(0.5)(attention)
    outputs = Dense(num_classes, activation='softmax')(dropout)

    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# Assuming you have train_df, dev_df, test_df
# 1. Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts(train_df['Text'])
vocab_size = len(tokenizer.word_index) + 1

# 2. Convert Text to Sequences
train_sequences = tokenizer.texts_to_sequences(train_df['Text'])
dev_sequences = tokenizer.texts_to_sequences(dev_df['Text'])
test_sequences = tokenizer.texts_to_sequences(test_df['Text'])

# 3. Padding
max_sequence_length = max(len(seq) for seq in train_sequences)
train_padded = pad_sequences(train_sequences, maxlen=max_sequence_length, padding='post')
dev_padded = pad_sequences(dev_sequences, maxlen=max_sequence_length, padding='post')
test_padded = pad_sequences(test_sequences, maxlen=max_sequence_length, padding='post')

# 4. Label Encoding
label_encoder = LabelEncoder()
train_labels_encoded = label_encoder.fit_transform(train_df['Label'])
dev_labels_encoded = label_encoder.transform(dev_df['Label'])
test_labels_encoded = label_encoder.transform(test_df['Label'])

# Build and train the model
model = build_bilstm_with_attention(vocab_size, 100, max_sequence_length, len(label_encoder.classes_))
model.fit(train_padded, train_labels_encoded, epochs=10, batch_size=32, validation_data=(dev_padded, dev_labels_encoded))

# Evaluate the model
predictions = model.predict(test_padded)
predicted_labels = np.argmax(predictions, axis=1)
print(classification_report(test_labels_encoded, predicted_labels, target_names=label_encoder.classes_))
```

```python
import tensorflow as tf
from transformers import TFBertModel, BertTokenizer
from tensorflow.keras.layers import Layer, Input, Dense, Dropout
from tensorflow.keras.models import Model
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
import numpy as np


bert_model_name = 'bert-base-multilingual-cased'
bert_tokenizer = BertTokenizer.from_pretrained(bert_model_name)
bert_model = TFBertModel.from_pretrained(bert_model_name)


max_seq_len = 128


class BertLayer(Layer):
    def __init__(self, **kwargs):
        super(BertLayer, self).__init__(**kwargs)
        self.bert = bert_model





    def call(self, inputs):
        input_ids, attention_mask = inputs
        output = self.bert(input_ids, attention_mask=attention_mask)[0]
        return output

def build_mbert_classifier(num_classes):
    input_ids = Input(shape=(max_seq_len,), dtype=tf.int32, name="input_ids")
    attention_mask = Input(shape=(max_seq_len,), dtype=tf.int32, name="attention_mask")

    bert_output = BertLayer()([input_ids, attention_mask])
    pooled_output = tf.keras.layers.GlobalAveragePooling1D()(bert_output)

    dropout = Dropout(0.5)(pooled_output)
    outputs = Dense(num_classes, activation='softmax')(dropout)

    model = Model(inputs=[input_ids, attention_mask], outputs=outputs)
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

```
